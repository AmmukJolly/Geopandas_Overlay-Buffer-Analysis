**Overlay Analysis**

Union : Combines all geometries from both layers

```python
data_union=gpd.overlay(data1,data2,how='union')

ax,extent=plt.subplots(figsize=(10,10))
data1.plot(ax=extent,color='white',edgecolor='purple')
data2.plot(ax=extent,color='pink')
data_union.plot(figsize=(10,10),ax=extent,color='orange')
plt.title("Union")
```

Intersection:  Keeps only the area where both layers overlap.

```python
data_intersection=gpd.overlay(data1,data2,how='intersection')

ax,extent=plt.subplots(figsize=(10,10))
data1.plot(ax=extent,color='white',edgecolor='purple')
data2.plot(ax=extent,color='pink')
data_intersection.plot(figsize=(10,10),ax=extent,color='orange')
plt.title("Intersection")
```

Identity : : Keeps all of A, and adds attributes of B wherever they intersect.

```python
ax,extent=plt.subplots(figsize=(10,10))
data1.plot(ax=extent,color='white',edgecolor='purple')
data2.plot(ax=extent,color='pink')


data_identity=gpd.overlay(data1,data2,how='identity')
data_identity.plot(figsize=(10,10),ax=extent,color='red')
plt.title("Identity")
```

Difference : Keeps the part of A that is NOT overlapping with B.



```python
ax,extent=plt.subplots(figsize=(10,10))
data1.plot(ax=extent,color='white',edgecolor='purple')
data2.plot(ax=extent,color='pink')


data_diff=gpd.overlay(data1,data2,how='difference')
data_diff.plot(ax=extent,color='blue')
plt.title("Difference")
```

Symmetric Difference: Keeps areas in A or B but not where they overlap.

```python
ax,extent=plt.subplots(figsize=(10,10))
data1.plot(ax=extent,color='white',edgecolor='purple')
data2.plot(ax=extent,color='pink')


data_symdiff=gpd.overlay(data1,data2,how='symmetric_difference')
plt.title("Symmetric Difference")
data_symdiff.plot(ax=extent,color='green')
```

Buffer Analysis:


Buffer analysis is the process of drawing a boundary at a set distance around a map feature to find out whats nearby or affected.

```python
import geopandas as gpd
```

```python
hosp_loc=gpd.read_file("/content/HospitalsLoc.shp")

```

```python
hosp_loc
```

To get the general information - .info()

```python
hosp_loc.info()
```

To check the crs - .crs

```python
hosp_loc.crs
```

To convert the GCS into PCS - .to_crs(epsg= )

```python
Main_hosp_loc =hosp_loc.to_crs(epsg=32643)
```

```python
Main_hosp_loc.crs
```

To save the new .shp file - .to_file(" ")

```python
Main_hosp_loc.to_file("/content/sample_data/Buffer/new_utm_hosp_loc.shp")
```

```python
bufferdata=Main_hosp_loc.buffer(500)
bufferdata.plot()
```

```python
import matplotlib.pyplot as plt
ax,extent=plt.subplots(figsize=(10,25))

bufferdata.plot(ax=extent,color='white',edgecolor="green")
Main_hosp_loc.plot(ax=extent,color='red')
plt.title("Buffer Data")
```
